## bubble_sort

```python
def bubble_sort(array:list):
    
    if len(array) < 2:
        return array
    
    idx = 0
    idx_last = len(array) - 1
    
    while idx_last:
        if idx + 1 <= idx_last:
            if array[idx] > array[idx+1]:
                array[idx], array[idx+1] = array[idx+1], array[idx]
            idx += 1
        else:
            idx_last -= 1
            idx = 0
    return array
```

## Radix Sort (only can be used in number)

```python
def max_(array:list):
    if not array:
        raise ValueError("array can't be empty")
    MAX =  - float("inf")
    for i in array:
        if i > MAX:
            MAX = i
    return MAX

def split_negitive(array):
    return list(filter(lambda x: x >= 0, array)), list(filter(lambda x: x < 0, array))

def digit_counts(num):
    i = 0
    while num:
        num //= 10
        i += 1
    return i
```

```python
def get_digit(i, num:int):
    try:
        n = str(num)[-i]
    except IndexError:
        return '0'
    return n

def radix_sort(array:list):
    if not array:
        return []
    m = digit_counts(max_(array))
    for i in range(1, m+1):
        number_mappings = {str(num): list() for num in range(10)}
        keys = list(map(lambda x: get_digit(i, x), array))
        for idx, key in enumerate(keys):
            number_mappings[key].append(array[idx])
        array = [i for key, item in number_mappings.items() for i in item]
    return array
```

```python
def radix_sort_extend(array):
    a, b = split_negitive(array)
    b = list(map(lambda x: x * -1, b))
    a = radix_sort(a)
    b = radix_sort(b)
    b = list(map(lambda x: x * -1, b))
    b = list(reversed(b))
    return b + a
```

```python
radix_sort_extend(test_array)
```
